```python
#Initializations
import os

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import rc
from sklearn.metrics import r2_score

mpl.rcParams.update(mpl.rcParamsDefault)
mpl.rcParams.update({'mathtext.fontset': 'cm'})
mpl.rcParams.update({'font.family':'serif'})
mpl.rcParams.update({'font.serif':'STIXGeneral'})

mpl.rcParams.update({'axes.titlesize': 20})
mpl.rcParams.update({'axes.labelsize': 24})
mpl.rcParams.update({'axes.linewidth': 0.5})
mpl.rcParams.update({'xtick.labelsize': 16})
mpl.rcParams.update({'ytick.labelsize': 16})
mpl.rcParams.update({'legend.framealpha': 1})
mpl.rcParams.update({'axes.titlepad': 12})
mpl.rcParams.update({'figure.dpi': 100})

plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['axes.labelsize'] = 22
plt.rcParams['axes.titlesize'] = 14
plt.rcParams["font.family"] = "Times New Roman"
plt.clf()
rc('text', usetex=True)

N_closest = 50
Re = 10
phi = 10
shape = "cube"
target = "Drag"

if phi == 5:
    phi_sys = "0p05"
    N_par = 1146
elif phi == 10:
    phi_sys = "0p1"
    N_par = 2292
else:
    phi_sys = "0p2"
    N_par = 2197
my_path = "/home/zihao/Documents/PhD/Post-Processing/3D/Bidisperse/Platonic_Solids/"+shape+f"_Re_{Re}_phi_"+phi_sys+f"_M_{N_closest}"
os.chdir(my_path)

fig = plt.figure(figsize=(12, 4))

SCATTER_COLOR = 'cornflowerblue'
ALPHA = 0.5
SMALL_LABEL = 16
LINEWIDTH = 1

D_in = np.load(target+"_true.npy")
D_pred = np.load(target+"_pred.npy")

# generate the best fit for the predicted data ___________________
lim1 = np.mean(D_in) - 3.5*np.std(D_in)
lim2 = np.mean(D_in) + 3.5*np.std(D_in)
lim3 = np.mean(D_pred) - 2*np.std(D_pred)

# # Plots _____________________
ax1 = plt.subplot(131)

ax1.scatter(D_in, D_pred,
            marker='o',
            s=10, linewidth=0.4,
            alpha=0.5,
            edgecolor='black', facecolor='white')
ax1.plot([lim1, lim2], [lim1, lim2], linestyle='-', color='blue', linewidth=2.5, alpha=0.3)
#     ax1.plot(D_vec, D_reg2, linestyle='-', color='black', linewidth=2.5, alpha=0.1)

ax1.set_aspect('equal')

coord = 'F_x'
if coord != None:
    if coord == 'F_x':
        ax1.set_xlabel(r'$\tilde{F}_{x,DNS}$', labelpad=15)
        ax1.set_ylabel(r'$\tilde{F}_{x,PINN}$', labelpad=10)
    elif coord == 'F_y':
        ax1.set_xlabel(r'$\tilde{F}_{y,DNS}$', labelpad=15)
        ax1.set_ylabel(r'$\tilde{F}_{y,PINN}$', labelpad=10)
    elif coord == 'T_z':
        ax1.set_xlabel(r'$\tilde{T}_{z,DNS}$', labelpad=15)
        ax1.set_ylabel(r'$\tilde{T}_{z,PINN}$', labelpad=10)

ax1.set_xlim((1.0*lim1, 1.0*lim2))
ax1.set_ylim((1.0*lim1, 1.0*lim2))

#     ax1.set_xticks(range(-4,5,1))
#     ax1.set_yticks(range(-4,5,1))

# R^2 on plot _____________________
x_ann = ax1.get_xlim()[1] - 0.05*(plt.xlim()[1]-plt.xlim()[0])
y_ann = ax1.get_ylim()[0] + 0.05*(plt.ylim()[1]-plt.ylim()[0])
ax1.annotate(str(r'$\mathrm{R^2} = ' + f'{r2_score(D_in, D_pred):.2f}$'),
             xy=(x_ann, y_ann),
             horizontalalignment='right',
             verticalalignment='bottom',
             fontsize=18);

# ticks _____________________
from matplotlib.ticker import AutoMinorLocator, MaxNLocator, MultipleLocator

ax1.xaxis.set_major_locator(MaxNLocator(5))
ax1.xaxis.set_minor_locator(AutoMinorLocator(2))

ax1.yaxis.set_major_locator(MaxNLocator(5))
ax1.yaxis.set_minor_locator(AutoMinorLocator(2))

ax1.tick_params(which='major', length=5)
ax1.tick_params(which='minor', length=2)

# Set the font name for axis tick labels _____________________
for tick in ax1.get_xticklabels():
    tick.set_fontname('Times New Roman')
for tick in ax1.get_yticklabels():
    tick.set_fontname('Times New Roman')
    
#==============================================================

target = "Lift"

D_in = np.load(target+"_true.npy")
D_pred = np.load(target+"_pred.npy")

# generate the best fit for the predicted data ___________________
lim1 = np.mean(D_in) - 3.5*np.std(D_in)
lim2 = np.mean(D_in) + 3.5*np.std(D_in)
lim3 = np.mean(D_pred) - 2*np.std(D_pred)

# Plots _____________________
ax2 = plt.subplot(132)

ax2.scatter(D_in, D_pred,
            marker='o',
            s=10, linewidth=0.4,
            alpha=0.5,
            edgecolor='black', facecolor='white')
ax2.plot([lim1, lim2], [lim1, lim2], linestyle='-', color='blue', linewidth=2.5, alpha=0.3)
#     ax1.plot(D_vec, D_reg2, linestyle='-', color='black', linewidth=2.5, alpha=0.1)

ax2.set_aspect('equal')

coord = 'F_y'
if coord != None:
    if coord == 'F_x':
        ax2.set_xlabel(r'$\tilde{F}_{x,DNS}$', labelpad=15)
        ax2.set_ylabel(r'$\tilde{F}_{x,PINN}$', labelpad=10)
    elif coord == 'F_y':
        ax2.set_xlabel(r'$\tilde{F}_{y,DNS}$', labelpad=15)
        ax2.set_ylabel(r'$\tilde{F}_{y,PINN}$', labelpad=10)
    elif coord == 'T_z':
        ax2.set_xlabel(r'$\tilde{T}_{z,DNS}$', labelpad=15)
        ax2.set_ylabel(r'$\tilde{T}_{z,PINN}$', labelpad=10)

ax2.set_xlim((1.0*lim1, 1.0*lim2))
ax2.set_ylim((1.0*lim1, 1.0*lim2))

#     ax1.set_xticks(range(-4,5,1))
#     ax1.set_yticks(range(-4,5,1))

# R^2 on plot _____________________
x_ann = ax2.get_xlim()[1] - 0.05*(plt.xlim()[1]-plt.xlim()[0])
y_ann = ax2.get_ylim()[0] + 0.05*(plt.ylim()[1]-plt.ylim()[0])
ax2.annotate(str(r'$\mathrm{R^2} = ' + f'{r2_score(D_in, D_pred):.2f}$'),
             xy=(x_ann, y_ann),
             horizontalalignment='right',
             verticalalignment='bottom',
             fontsize=18);

# ticks _____________________
from matplotlib.ticker import AutoMinorLocator, MaxNLocator, MultipleLocator

ax2.xaxis.set_major_locator(MaxNLocator(5))
ax2.xaxis.set_minor_locator(AutoMinorLocator(2))

ax2.yaxis.set_major_locator(MaxNLocator(5))
ax2.yaxis.set_minor_locator(AutoMinorLocator(2))

ax2.tick_params(which='major', length=5)
ax2.tick_params(which='minor', length=2)

# Set the font name for axis tick labels _____________________
for tick in ax2.get_xticklabels():
    tick.set_fontname('Times New Roman')
for tick in ax2.get_yticklabels():
    tick.set_fontname('Times New Roman')
    
#==============================================================

target = "Torque"

D_in = np.load(target+"_true.npy")
D_pred = np.load(target+"_pred.npy")

# generate the best fit for the predicted data ___________________
lim1 = np.mean(D_in) - 3.5*np.std(D_in)
lim2 = np.mean(D_in) + 3.5*np.std(D_in)
lim3 = np.mean(D_pred) - 2*np.std(D_pred)

# Plots _____________________
ax3 = plt.subplot(133)

ax3.scatter(D_in, D_pred,
            marker='o',
            s=10, linewidth=0.4,
            alpha=0.5,
            edgecolor='black', facecolor='white')
ax3.plot([lim1, lim2], [lim1, lim2], linestyle='-', color='blue', linewidth=2.5, alpha=0.3)
#     ax1.plot(D_vec, D_reg2, linestyle='-', color='black', linewidth=2.5, alpha=0.1)

ax3.set_aspect('equal')

coord = 'T_z'
if coord != None:
    if coord == 'F_x':
        ax3.set_xlabel(r'$\tilde{F}_{x,DNS}$', labelpad=15)
        ax3.set_ylabel(r'$\tilde{F}_{x,PINN}$', labelpad=10)
    elif coord == 'F_y':
        ax3.set_xlabel(r'$\tilde{F}_{y,DNS}$', labelpad=15)
        ax3.set_ylabel(r'$\tilde{F}_{y,PINN}$', labelpad=10)
    elif coord == 'T_z':
        ax3.set_xlabel(r'$\tilde{T}_{z,DNS}$', labelpad=15)
        ax3.set_ylabel(r'$\tilde{T}_{z,PINN}$', labelpad=10)

ax3.set_xlim((1.0*lim1, 1.0*lim2))
ax3.set_ylim((1.0*lim1, 1.0*lim2))

#     ax1.set_xticks(range(-4,5,1))
#     ax1.set_yticks(range(-4,5,1))

# R^2 on plot _____________________
x_ann = ax3.get_xlim()[1] - 0.05*(plt.xlim()[1]-plt.xlim()[0])
y_ann = ax3.get_ylim()[0] + 0.05*(plt.ylim()[1]-plt.ylim()[0])
ax3.annotate(str(r'$\mathrm{R^2} = ' + f'{r2_score(D_in, D_pred):.2f}$'),
             xy=(x_ann, y_ann),
             horizontalalignment='right',
             verticalalignment='bottom',
             fontsize=18);

# ticks _____________________
from matplotlib.ticker import AutoMinorLocator, MaxNLocator, MultipleLocator

ax3.xaxis.set_major_locator(MaxNLocator(5))
ax3.xaxis.set_minor_locator(AutoMinorLocator(2))

ax3.yaxis.set_major_locator(MaxNLocator(5))
ax3.yaxis.set_minor_locator(AutoMinorLocator(2))

ax3.tick_params(which='major', length=5)
ax3.tick_params(which='minor', length=2)

# Set the font name for axis tick labels _____________________
for tick in ax3.get_xticklabels():
    tick.set_fontname('Times New Roman')
for tick in ax3.get_yticklabels():
    tick.set_fontname('Times New Roman')
    
fig.tight_layout()

pos = ax2.get_position()
new_pos = [pos.x0-0.01, pos.y0, pos.width, pos.height]
ax2.set_position(new_pos)

plt.show()
file_name = "PINN_correlation_"+shape+f"_Re_{Re}_phi_"+phi_sys+".png"
fig.savefig(file_name, bbox_inches="tight", dpi=200)
```
